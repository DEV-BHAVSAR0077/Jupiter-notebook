```python
import pickle

# Step 1: Create a Python list
my_list = ["apple", "banana", "cherry", "mango"]
print("Original List:", my_list)

# Step 2: Write the list to a text file (each element on new line)
with open("my_list.txt", "w") as f:
    for item in my_list:
        f.write(item + "\n")

print("List written to my_list.txt")
```

```python
# Step 3: Read the text file and reconstruct the list
loaded_list = []
with open("my_list.txt", "r") as f:
    for line in f:
        loaded_list.append(line.strip())

print("List loaded from text file:", loaded_list)

# Step 4: Convert (Serialize) the list into a pickle file
with open("my_list.pkl", "wb") as pkl_file:
    pickle.dump(loaded_list, pkl_file)

print("List serialized and saved as my_list.pkl")
```

```python
# Step 5: Load back the pickle file to verify
with open("my_list.pkl", "rb") as pkl_file:
    final_list = pickle.load(pkl_file)

print("List loaded from pickle file:", final_list)
```

```python
import pandas as pd
import numpy as np

# Number of rows
n = 100000

# Creating DataFrame with 3 columns
df = pd.DataFrame({
    'Column1': np.random.randint(1, 1000000, n),
    'Column2': np.random.rand(n),
    'Column3': np.random.choice(['A', 'B', 'C', 'D'], n)
})

# Save as CSV file
df.to_csv("my_large_dataframe.csv", index=False)

print("CSV file created successfully with 100000 rows and 3 columns!")
```
